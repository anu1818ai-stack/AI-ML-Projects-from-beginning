a simple calculator using funtion




```python
# step 1: funtion for each operation
def add(a,b):
  return a + b

def sub(a,b):
  return a - b

def multiply(a,b):
  return a * b

def divide(a,b):
  return a / b

def power(a,b):
  return a ** b

def mod(a,b):
  return a % b

def floor(a,b):
  return a // b
#step2: selet option
print("Select operation:")
print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")
print("5. Power")
print("6. Mod")
print("7. Floor")
#step4: user input
user_choice = input("Enter choice (1/2/3/4/5/6/7): ")
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

#step5: Calculator performs
if user_choice == "1":
  print("Result=", add(num1, num2))
elif user_choice == "2":
  print("Result=", sub(num1, num2))
elif user_choice == "3":
  print("Result=", multiply(num1, num2))
elif user_choice == "4":
  print("Result=", divide(num1, num2))
elif user_choice == "5":
  print("Result=", power(num1, num2))
elif user_choice == "6":
  print("Result=", mod(num1, num2))
elif user_choice == "7":
  print("Result=", floor(num1, num2))
else:
  print("Invalid input")
```

A simple calculator using lambda funtion for shorten the code â–ª

```python
# define lambda funtion
a = lambda a,b: a + b
b = lambda a,b: a - b
c = lambda a,b: a * b
d = lambda a,b: a / b
e = lambda a,b: a ** b
f = lambda a,b: a % b
g = lambda a,b: a // b

# selet option
print("Select operation:")
print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")
print("5. Power")
print("6. Mod")
print("7. Floor")

# user input
user_choice = input("Enter choice (1/2/3/4/5/6/7): ")
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

# calculator performs
if user_choice == "1":
  print("Result=", a(num1, num2))
elif user_choice == "2":
  print("Result=", b(num1, num2))
elif user_choice == "3":
  print("Result=", c(num1, num2))
elif user_choice == "4":
  print("Result=", d(num1, num2))
elif user_choice == "5":
  print("Result=", e(num1, num2))
elif user_choice == "6":
   print("Result=", f(num1, num2))
elif user_choice == "7":
    print("Result=", g(num1, num2))
else:
  print("Invalid input")


```
